```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("salesforcourse-4fe2kehu.csv")
df
```

```python
df.info()
```

### Year Analysis

```python
year_count=df['Year'].value_counts()
year_percent=df['Year'].value_counts(normalize=True)*100
freq_table1=pd.DataFrame({'Frequency':year_count,'Percentage %':year_percent})
freq_table1
```

```python
plt.pie(year_count,labels=['2016','2015'],colors=['cyan','purple'],shadow=True,autopct='%1.1f%%')
plt.legend()
plt.show()
```

- 2016 is the most year of transaction occured 

### Month Analysis

```python
month_count=df['Month'].value_counts()
month_percent=df['Month'].value_counts(normalize=True)*100
freq_table2=pd.DataFrame({'Frequency':month_count,'Percentage %':month_percent})
freq_table2
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Month',data=df)
plt.show()
```

- June & May are the most months of transaction occured

```python
country_count=df['Country'].value_counts()
country_percent=df['Country'].value_counts(normalize=True)*100
freq_table4=pd.DataFrame({'Frequency':country_count,'Percentage %':country_percent})
freq_table4
```

```python
cat_count=df['Product Category'].value_counts()
cat_percent=df['Product Category'].value_counts(normalize=True)*100
freq_table5=pd.DataFrame({'Frequency':cat_count,'Percentage %':cat_percent})
freq_table5
```

### Correlation 

```python
plt.subplot(2, 2, 1)
plt.scatter(df['Unit Price'], df['Revenue'])
plt.title('Unit Price vs. Revenue')
plt.xlabel('Unit Price')
plt.ylabel('Revenue')


plt.subplot(2, 2, 2)
plt.scatter(df['Unit Cost'], df['Cost'])
plt.title('Unit Cost vs. Cost')
plt.xlabel('Unit Cost')
plt.ylabel('Cost')


plt.subplot(2, 2, 3)
plt.scatter(df['Unit Price'], df['Cost'])
plt.title('Unit Price vs. Cost')
plt.xlabel('Unit Price')
plt.ylabel('Cost')


plt.subplot(2, 2, 4)
plt.scatter(df['Unit Cost'], df['Revenue'])
plt.title('Unit Cost vs. Revenue')
plt.xlabel('Unit Cost')
plt.ylabel('Revenue')

plt.tight_layout()
plt.show()
```

- Unit price vs Revenu higher unit prices generally correspond to higher revenues.
- Unit cost vs Cost strong positive linear relationship.
- Unit price vs Cost spread of data points, with a positive relationship.
- Unit price vs revenue positive relationship, but the data points are widely scattered.
